```python
df_selected = df.iloc[:, [0, 3, 9, 13]]
df_selected.head()
```

```python
# Remove the first row (header-like row)
df_selected_clean = df_selected.iloc[1:].reset_index(drop=True)

# Assign meaningful English column names
df_selected_clean.columns = ['year', 'avg_salary', 'v_idx', 'e_60']

# Export to JSON
df_selected_clean.to_json('df_selected_clean.json', orient='records', force_ascii=False)
```

```python
df_selected_clean
```

```python
import json

# Convert 'year' column to string to ensure it can be used as a JSON key
df_selected_clean['year'] = df_selected_clean['year'].astype(str)

# Set 'year' as index and convert to dictionary
json_dict = df_selected_clean.set_index('year').to_dict(orient='index')

with open('df_selected_clean_by_year.json', 'w', encoding='utf-8') as f:
    json.dump(json_dict, f, ensure_ascii=False, indent=2)

json_dict
```
